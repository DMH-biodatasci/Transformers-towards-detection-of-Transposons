## metrics of chunked data


```python
import pandas as pd
import os
import shelve
import numpy as np
```

```python
results = shelve.open(os.path.join("..", "data", "prep_trainvalidationtest_1"))
df_chunked = results["chunks"]
tokenize_table = results["tokenize_table"]
results.close
```

```python
print("number of files scanned: ", df_chunked["origin"].nunique())
print('number of chunks:',df_chunked.shape[0])
print("number of unique tokens:", len(tokenize_table.keys()))
print("\nnumber of na per column:")
print(df_chunked.isna().sum())
```

```python
for set in ["training","validation","test"]:
    n_t = np.array(df_chunked[df_chunked["set"] == set]["attention_masks"]).sum().sum()
    n_TE = np.array(df_chunked[df_chunked["set"] == set]["labels"]).sum().sum()
    
    s1 = 0
    s2 = 0
    for c in df_chunked[df_chunked["set"] == set]["labels"]:
        s2 = s2 + 1
        if sum(c) > 0:
         s1 = s1 + 1
  
    print("\n", set)
    print("  number of chunks: {:,d} = {:n}%".format(s2, 100))
    print("  number of chunks with TE tokens: {:,d} = {:n}%".format(s1, s1/s2*100))
    print("  number of tokens : {:,d} = {:n}%".format(n_t,100))
    print("  number of TE tokens: {:,d} = {:n}%".format(n_TE,n_TE/n_t*100))
```

```python
# get length of each chunk
l = []
for row in list(df_chunked["attention_masks"]):
  l.append(sum(row))
df_chunked["length"] = l

# get number of TE elements per chunk
e = []
for row in list(df_chunked["labels"]):
  e.append(sum(row))
df_chunked["TEs"] = e

print("length of chunk     number of chunks     number of chunks with TE marks")
ll = sorted(list(df_chunked["length"].unique()))
for i in ll:
    x = df_chunked[(df_chunked["length"] == i) & (df_chunked["TEs"] > 0)]["TEs"].count()
    n = df_chunked[df_chunked["length"] == i]["TEs"].count()
    #print(x)
    #print(n)
    print("     {:>6}            {:>6}            {:>6}".format(i, n, x))
```
